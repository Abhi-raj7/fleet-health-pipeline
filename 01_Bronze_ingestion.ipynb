```python
#data reading and writing to broze schema from the source

df_maintenance_logs =  spark.read.format('csv').option('header',True)\
                .option('inferSchema',False)\
                .load("/Volumes/proj_veh/raw/raw_data/maintenance logs.csv")

df_maintenance_logs.write.mode("overwrite").format('delta').saveAsTable("proj_veh.bronze.maintenance_logs")

#--------------------------------------------------------------------------

df_oem_recalls =  spark.read.format('csv').option('header',True)\
                .option('inferSchema',False)\
                .load("/Volumes/proj_veh/raw/raw_data/oem recalls.csv")

df_oem_recalls.write.mode("overwrite").format('delta').saveAsTable("proj_veh.bronze.oem_recalls")

#------------------------------------------------------------------------
df_vehicle_master =  spark.read.format('csv').option('header',True)\
                .option('inferSchema',False)\
                .load("/Volumes/proj_veh/raw/raw_data/vehicle master.csv")

df_vehicle_master.write.mode("overwrite").format('delta').saveAsTable("proj_veh.bronze.vehicle_master")

#------------------------------------------------------------------------

df_warranty_claims =  spark.read.format('csv').option('header',True)\
                .option('inferSchema',False)\
                .load("/Volumes/proj_veh/raw/raw_data/warranty claims.csv")

df_warranty_claims.write.mode("overwrite").format('delta').saveAsTable("proj_veh.bronze.warranty_claims")
```

```python
#checking the num of rows
def row_count(table_name):
    count = spark.sql(f"select count(*) as cnt from {table_name}").collect()[0]['cnt']
    print(f"total num of rows {table_name} :{count}")
    return count
```

```python
row_count("proj_veh.Bronze.maintenance_logs")
row_count("proj_veh.Bronze.oem_recalls")
row_count("proj_veh.Bronze.vehicle_master")
row_count("proj_veh.Bronze.warranty_claims")
```
